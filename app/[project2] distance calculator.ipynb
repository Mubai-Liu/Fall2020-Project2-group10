```python
import pandas as pd
from math import sin,cos, sqrt, atan2
import geopy.distance
```

```python
df = pd.read_csv('state_att.csv')
df.head()
```

```python
def cal(lat,lon):
    coords_des = (lat, lon)
    coords_ny = (40.71427, -74.00597)
    return geopy.distance.distance(coords_des, coords_ny).miles
```

```python
df['distance'] = list(map(cal,df['Longitude (N/S)'],-df['Latitude (E/W)']))
```
